### Are there different success rates over geographies?

```python
# I should be able to measure the success rates by 
# success rates = registration / referrals for each program
# 1. measure the overall success rates
# 2. measure the success rates by city, county, states
```

```python
df = raw_df
```

```python
df.columns
```

```python
#ihp
df['ihpSuccessRates'] = [0 if row_df['ihpReferrals'] == 0 else  row_df['ihpEligible'] / row_df['ihpReferrals'] for idx, row_df in df.iterrows()]
```

```python
#ihp amount given per registration
df['ihpAmountperRegistration'] = [0 if row_df['ihpEligible'] == 0 else  row_df['ihpAmount'] / row_df['ihpEligible'] for idx, row_df in df.iterrows()]
```

```python
#ha
df['haSuccessRates'] = [0 if row_df['haReferrals'] == 0 else  row_df['haEligible'] / row_df['haReferrals'] for idx, row_df in df.iterrows()]
```

```python
#ha
df['haAmountperRegistration'] = [0 if row_df['haEligible'] == 0 else  row_df['haAmount'] / row_df['haEligible'] for idx, row_df in df.iterrows()]
```

```python
#ona
df['onaSuccessRates'] = [0 if row_df['onaReferrals'] == 0 else  row_df['onaEligible'] / row_df['onaReferrals'] for idx, row_df in df.iterrows()]
```

```python
#ona
df['onaAmountperRegistration'] = [0 if row_df['onaEligible'] == 0 else  row_df['onaAmount'] / row_df['onaEligible'] for idx, row_df in df.iterrows()]
```

```python
profile = ProfileReport(df, title="Wildfire Profiling Report", explorative=True)
profile.to_widgets()
```
